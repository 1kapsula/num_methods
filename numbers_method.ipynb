```python
import numpy as np

def bin_method(poly, l, r, eps):
    while (r - l) / 2 > eps:
        m = (l + r) / 2
        if poly(m) == 0:
            return m
        elif poly(l) * poly(m) < 0:
            r = m
        else:
            l = m
    return (l + r) / 2

values = input("Введите коэффициенты уравнения через пробел: ")
coeff = [float(val) for val in values.split()]
eps = float(input("Введите отклонение: "))

poly = np.poly1d(coeff)

roots = bin_method(poly,-10, 12, eps)

print(roots)
```

def dihot(degree, l, edge_positiv, kf, eps):
    while (-edge_positiv - l) / 2 > eps:
        x = 0.5 * (l + edge_positiv)
        if x == l or x == edge_positiv:
            return x
        if polinom(degree, x, kf) < 0:
            l = x
        else:
            edge_positiv = x


def step_up(level, A, B, current_roots_count, eps):
    major = 0
    for i in range(level):
        s = abs(A[level][i])
        if s > major:
            major = s
    major += 1.0

    current_roots_count[level] = 0

    for i in range(current_roots_count[level - 1] + 1):
        if i == 0:
            edge_left = -major
        else:
            edge_left = B[level - 1][i - 1]

        rb = polinom(level, edge_left, A[level])

        if rb == 0:
            B[level][current_roots_count[level]] = edge_left
            current_roots_count[level] += 1
            continue

        sign_left = 1 if rb > 0 else -1

        if i == current_roots_count[level - 1]:

            edge_negativ = edge_right

        B[level][current_roots_count[level]] = dihot(level, edge_negativ, edge_positiv, A[level], eps)

```python
def polinom(n, x, k):
    s = 1
    for i in range(n-1, -1, -1):
        s = s * x + k[i]
    return s

def dihot(degree, l, r, kf, eps):
    while abs(l - r) > eps:
        x = (l + r) / 2
        p_l = polinom(degree, l, kf)
        p_x = polinom(degree, x, kf)
        if p_x == 0:
            return x
        if p_l * p_x < 0:
            r = x
        else:
            l = x
    return x

def step_up(level, A, B, curr_cnt, eps):
    m = 0
    for i in range(level):
        s = abs(A[level][i])
        if s > m:
            m = s
    m += 1.0

    curr_cnt[level] = 0

    for i in range(curr_cnt[level - 1] + 1):
        if i == 0:
            edge_left = -m
        else:
            edge_left = B[level - 1][i - 1]

        rb = polinom(level, edge_left, A[level])

        if rb == 0:
            B[level][curr_cnt[level]] = edge_left
            curr_cnt[level] += 1
            continue

        sign_left = 1 if rb > 0 else -1

        if i == curr_cnt[level - 1]:
            edge_right = m
        else:
            edge_right = B[level - 1][i]

        rb = polinom(level, edge_right, A[level])
        
        if rb == 0:
            B[level][curr_cnt[level]] = edge_right
            curr_cnt[level] += 1
            continue

        sign_right = 1 if rb > 0 else -1

        if sign_left == sign_right:
            continue

        if sign_left < 0:
            l = edge_left
            r = edge_right
        else:
            l = edge_right
            r = edge_left

        B[level][curr_cnt[level]] = dihot(level, l, r, A[level], eps)
        curr_cnt[level] += 1

def polynom_real_roots(n, kf, eps):
    A = [[0.0]*(i+1) for i in range(n+1)]
    B = [[0.0]*(i+1) for i in range(n+1)]
    curr_cnt = [0] * (n + 1)
    

    for i in range(n+1):
        A[n][i] = kf[i] / kf[n]

    for i1, i in zip(range(n, 0, -1), range(n-1, -1, -1)):
        for j1, j in zip(range(i+1, -1, -1), range(i, -1, -1)):
            A[i][j] = A[i1][j1] * j1

    for i in range(1, n):
        for j in range(len(A[i])):
            A[i][j] /= A[i][-1]

    curr_cnt[1] = 1
    
    B[1][0] = -A[1][0]
    for i in range(2, n+1):
        step_up(i, A, B, curr_cnt, eps)
    roots_cnt = curr_cnt[n]
    roots_ans = B[n][:roots_cnt]

    return roots_ans, roots_cnt

def main():
    n = int(input("Введите степень полинома: "))

    kf = [0.0] * (n + 1)
    print("Введите коэффициенты полинома (через пробел начина от наименьшей степени):")
    kf = list(map(float, input().split()))
    eps = float(input("Введите отклонение: "))
    roots_ans, roots_cnt = polynom_real_roots(n, kf, eps)

    print(f"Количество найденных корней: {roots_cnt}")
    if roots_cnt > 0:
        print("Корни полинома:")
        for root in roots_ans:
            print(f"{root:.{len(str(eps)) - 2}f}")

if __name__ == "__main__":
    main()
```
